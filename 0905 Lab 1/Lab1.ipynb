```python
import numpy as np
import pandas as pd

# numpy example: create an array and calculate sum and mean
arr = np.array([1, 2, 3, 4, 5])
arr_sum = np.sum(arr)
arr_mean = np.mean(arr)
print(f"Array: {arr}")
print(f"Sum: {arr_sum}, Mean: {arr_mean}")
# pandas example: create a DataFrame and display basic statistics
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'Los Angeles', 'Chicago']}
df = pd.DataFrame(data)
print("\nDataFrame:")
print(df)
print("\nDataFrame Statistics:")
print(df.describe())    
# This script demonstrates basic usage of numpy and pandas libraries. 
```
